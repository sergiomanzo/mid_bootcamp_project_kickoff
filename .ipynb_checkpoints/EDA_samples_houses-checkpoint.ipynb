Orientation description in Madrid   



#South facing house: 
The sun is direct in this orientation during the whole day except in summer, where it will affect the central hours of the day.

#North-facing house: 
The sun only shines in the first hours of the morning and the last hours of the night. You will enjoy more sun in summer, although not directly, and it is a good orientation to maintain cool areas in the house.

#South facing house: 
The sun is direct in this orientation during the whole day except in summer, where it will affect in the central hours of the day.
It is the orientation par excellence to maintain a good temperature in the entrance of the house or a small garden. During the summer, it is advisable to use awnings to protect the facade from the sun and reduce the cost of air conditioning.

#East-facing house : 
The sun is direct from first thing in the morning until noon during almost the whole year.
It is the best orientation for a house in Madrid, since it is the orientation that involves a more balanced energy expenditure in both winter and summer.

#West facing house :
The sun behaves in the opposite way to the previous orientation. That is to say, it hits from noon to sunset.
It is not the best orientation for a house in any room, since it does not accumulate heat during the day in winter, while in summer it receives direct sun in the hottest hours.

```python
def spearman(frame, features):
    spr = pd.DataFrame()
    spr['feature'] = features
    spr['spearman'] = [frame[f].corr(frame['price'], 'spearman') for f in features]
    spr = spr.sort_values('spearman')
    plt.figure(figsize=(6, 0.25*len(features)))
    sns.barplot(data=spr, y='feature', x='spearman', orient='h')
    
features = quantitative + qualitative
spearman(pangeldf, features)
```

# Hypotesis testing

# Price Hypotesis

```python
#1 - Set the hypothesis
# I want appmts with at least 2 rooms and Terrace
# H0: The avg price in Neigh  Comillas <= The avg price in Neigh  P Angel
# H1: The avg price in Neigh Comillas > The avg price in Neigh  P Angel
#a  1 tail test, needs to use alternative = gater
```

```python
#STEP 3:
    #Getting the subsample
    
comillastest = housesdf['price'][(housesdf['rooms'] >= 2) & (housesdf['Terrace'] == 1)  & (housesdf['location'] == 'Comillas')  ]
pangeltest = housesdf['price'][(housesdf['rooms'] >= 2) & (housesdf['Terrace'] == 1)    & (housesdf['location'] == 'Puerta_angel')]


display('Comillas sub sample',comillastest.count())
display('P Angel sub sample',pangeltest.count())
```

```python
#STEP 4:

    #Compute Statistic:
meanco = comillastest.mean()
stdco = comillastest.std()

meanpa = pangeltest.mean()


display(meanco)
display(stdco)

display(meanpa)


stat = (meanco -meanpa)/(stdco/ np.sqrt(24))
stat
```

```python

#STEP 5:

    #Getting the p-value
    #p-value 1-tailed
    
p_value = st.t.sf(abs(stat),24-1)
display(p_value)
```

```python
st.ttest_1samp(comillastest,meanpa,alternative="greater")
```

```python
p_value1 =st.ttest_1samp(comillastest,meanpa,alternative="greater").pvalue
p_value1
```

```python
#Result
if p_value1 > 0.05:
    print("With the available data we dont have enoght evidence to reject the Null hypotesis H0")
elif p_value1 < 0.05:
    print('We reject the H0 hypotesis')
```

```python
#Result
if p_value > 0.05:
    print("With the available data we dont have enoght evidence to reject the Null hypotesis H0")
elif p_value < 0.05:
    print('We reject the H0 hypotesis')
```


Result:

We connot reject the  null hypotesis  H0: The avg price in Neigh  Comillas <= The avg price in Neigh  P Angel for apmts with 2 rooms and terrace.
