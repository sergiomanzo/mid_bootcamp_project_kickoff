```python
#  pangeldata 
#   comillasdata
```

```python
pangeldf = pangeldata
comillasdf = comillasdata
```

```python
#Definition for a function to clean num columns
def clean_num(row):
    if type(row) != int:
            row = row.replace(',','')
            row = row.replace('.','')
            row = re.findall(r'[0-9]+',row)
            row = row[0]
    return row
```

```python
pangeldf.head(2)
```

```python
#Apllying function to clean the 2 df for 6 colums
pangeldf['price'] = pangeldf['price'].apply(clean_num).astype(int)
pangeldf['price/m2'] = pangeldf['price/m2'].apply(clean_num).astype(int)
pangeldf['M2 built'] = pangeldf['M2 built'].apply(clean_num).astype(int)

comillasdf['price'] = comillasdf['price'].apply(clean_num).astype(int)
comillasdf['price/m2'] = comillasdf['price/m2'].apply(clean_num).astype(int)
comillasdf['M2 built'] = comillasdf['M2 built'].apply(clean_num).astype(int)

```

```python
comillasdf['price']
```

```python
pangeldf.sample(3)
```

```python
#Defining a fuction to convert all yes or no to 0 and 1
def yesno_to_01(row):
    if row == 'yes':
        row = 1
    else:
        row = 0
    return row
```

```python
pangeldf['Terrace'] = pangeldf['Terrace'].apply(yesno_to_01).astype(int)
pangeldf['Elevator'] = pangeldf['Elevator'].apply(yesno_to_01).astype(int)
pangeldf['garage'] = pangeldf['garage'].apply(yesno_to_01).astype(int)

comillasdf['Terrace'] = comillasdf['Terrace'].apply(yesno_to_01).astype(int)
comillasdf['Elevator'] = comillasdf['Elevator'].apply(yesno_to_01).astype(int)
comillasdf['garage'] = comillasdf['garage'].apply(yesno_to_01).astype(int)
```

```python
#Defining a fuction to convert all to a simpler location
def locationsimp(row):
    if row == 'Puerta del Ángel, Madrid ':
        row = 'Puerta_angel'
    elif row == 'puerta del Ángel, Madrid ':
        row = 'Puerta_angel'
    elif row == 'Comillas, Madrid':
        row = 'Comillas'
        
    return row
```

```python
pangeldf['location'] = pangeldf['location'].apply(locationsimp)

comillasdf['location'] = comillasdf['location'].apply(locationsimp)
```
